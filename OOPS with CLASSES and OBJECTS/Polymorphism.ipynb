Polymorphism

It is a core concept in oops that allows object of different class to be treated as objects of common superclass. It provides a way to perform a single action in different forms. Polymorphism is typically achieved through method overriding interfaces

```python
class Animal:
    def speak(self):
        return "Sound of an animal"

class Dog(Animal):
    def speak(self):
        return "Woof!"

class Cat(Animal):
    def speak(self):
        return "Meow"
```

```python
dog= Dog()
dog.speak()
```

Polymorphism in abstract base classes

ABCs are used to define common methods for a group of related projects. they can enforce that dereived classes implement particular methods, promoting consistency across different implementations

```python
from abc import ABC, abstractclassmethod

class Vehicle(ABC):
    @abstractclassmethod
    def start_engine(self):
        pass

class Car(Vehicle):
    def start_engine(self):
        return "Car Engine started"
```
