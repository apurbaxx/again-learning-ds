- Encapsulation and abstraction are two fundamental principles of oops that help in designing robust, maintainable and reusable code. Encapsulation involves bundling data and method that operate on the data within a single unit, while abstraction involves hiding complex implementation details and exposing only necessary features

- Encapsulation is the concept of wrapping data and methods together as a single unit. It restricts direct access to some of the objects components, which is a measns of preventing accidental interference and misuse of data

```python
### Encapsulation with getter and setter methods
### Public, protected, private variables

class Person:
    def __init__(self, name, age):
        self.name = name #public variables
        self.age = age

person = Person("Apurba", 18)
print(person.name)
```

```python
class Person:
    def __init__(self, name, age, gender):
        self.__name = name #private variables
        self.__age = age
        self.gender = gender #public variable

```

```python
person = Person("Apurba", 18, "Male")
```

```python
print(person.__name) #Name will not get printed since it is a private variable
```

```python
class Person:
    def __init__(self, name, age, gender):
        self._name = name #protected variables -> Only its derived class can access it
        self._age = age
        self.gender = gender #public variable
        
class Employee(Person):
    def __init__(self, name, age, gender):
        super().__init__(name,age,gender)
```

```python
employee = Employee("Krish", 34, "Male")
print(employee.name)#Wont get printed
```

```python
##Encapsulation with getter and setter method]
class Person:
    def __init__(self, name, age):
        self.__name=name
        self.__age=age

    #getter method for name
    def get_name(self):
        return self.__name

    #setter method for name
    def set_name(self, name):
        self.__name=name

    def set_age(self,age):
        if age > 0:
            self.__age = age
        else:
            print("Age cannot be negative")
```

Abstraction

Abstraction is a concept of hiding the complex implementation details and showing only the necessary features of an object. This helps in reducing programming complexity and effort

```python
from abc import ABC, abstractclassmethod, abstractmethod


class Vehicle(ABC):
    def drive(self):
        print("This vehicle is used for driving")

    @abstractmethod
    def start_engine(self):
        pass

class Car(Vehicle):
    def start_engine(self):
        print("Car engine started")

def operate_vehicle(vehicle):
    vehicle.start_engine()


car = Car()
operate_vehicle(car)
```
