```python
#Assignment
def count_text_file(filename):
    with open(filename, "r") as file:
        lines = file.readlines()
        line_count = len(lines)
        word_count = sum(len(line.split()) for line in lines)
        char_count = sum(len(line) for line in lines)
    return line_count, word_count, char_count

file_name = "example.txt"
line_count, word_count, char_count = count_text_file(file_name)
print(f"Line count: {line_count}, Word count: {word_count}, Character count: {char_count}")
```

```python
with open("example.txt", "w+") as file: #Writing and then reading the file
    file.write("Hello World \n")
    file.write("2nd line")
    content = file.read()
    print(content)

    #Now to read the file we need to point the cursor to the 1st line of the file
    file.seek(0)
    content = file.read()
    print(content)
```
