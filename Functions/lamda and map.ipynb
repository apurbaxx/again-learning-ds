The filter function function constructs an iterator from elements of an iterable for which a function returns true. It is used to filter out items from a list(or any other iterable based on a condition

```python
def even(numb):
    if numb%2==0:
        return True

lst = [1,2,3,4,5,6,6,7,7,8,8,9,9]
list(filter(even, lst))
```

```python
#filter with a lambda function
numbers = [1,2,3,4,45,6]
greater_than_5 = list(filter(lambda x:x>5, numbers))
greater_than_5
```

```python
## Filter with a lambda function
even_and_greater_than_5 = list(filter(lambda x:x%2==0 and x>5, numbers))
```
